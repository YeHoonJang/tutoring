## 3. 판다스 내장 그래프 도구 활용
* 판다스는 matplotlib 의 일부 기능 지원


```python
# ./남북한발전전력량.xlsx 파일을 데이터 프레임으로 변환
df = pd.read_excel('./남북한발전전력량.xlsx')
df.head(10)
```

```python
# 남한, 북한 발전량 합계 데이터만 추출하여 df_ns 에 할당
# 행 인덱스 변경 index = ['South', 'North']
df_ns = df.iloc[[0,5],:]
df_ns.index = ['South', 'North']

df_ns.drop(['전력량 (억㎾h)', '발전 전력별'], axis=1, inplace=True)
df_ns
```

```python
df_ns.columns.map(int) # 열 이름의 자료형을 정수형으로 변경
```

### 선 그래프


```python
# 선 그래프 그리기
df_ns.plot()
```

```python
# 행, 열 전치하여 다시 그리고 tdf_ns에 할당
# week_02 참고 - 전치 df.transpose() / df.T
df_ns = df_ns.T
df_ns.plot()
```

### 막대 그래프


```python
# 막대 그래프 그리기
df_ns.plot(kind='bar')
```

### 히스토그램


```python
# 히스토그램 그리기
df_ns.plot(kind='hist')
```

### 산점도

```python
# read_csv() 함수로 df 생성
# 파일 : './auto-mpg.csv'
df = pd.read_csv('./auto-mpg.csv', header=None)
df.head()
```

```python
# 열 이름을 지정
cols = ['mpg','cylinders','displacement','horsepower','weight', 'acceleration','model year','origin','name']
df.columns=cols
# 2개의 열을 선택하여 산점도 그리기
# x= weight, y = mpg
df.plot(kind='scatter', x='weight', y='mpg')
```
